### Julian Date, a algorithm with some pitfalls (4 pt) <font color='red'><b>COURSEWORK</b></font>

In the online article: [A one-line algorithm for Julian Date](http://adsabs.harvard.edu/full/1983IAPPP..13...16F)
we find an old and very compact formula for programmable calculators to calculate a Julian Date given a day, month and year.   
A Julian Date is in fact a
day number corresponding to the number of elapsed days since the beginning of a cycle of 7980 years.
That makes it a useful tool to calculate how many days there are between two dates.
The one-line algorithm is valid for the period that started in 1582 with the Gregorian calendar.
The formula from the article is:


``JD = 367*Y -7*(Y+(M+9)/12)/4 - 3*((Y+(M-9)/7)/100 + 1)/4 + (275*M)/9 + D + 1721029-0.5``

The article notes: **It is important to note that after all divisions in this expression only that portion of the
quotient to the left of the decimal point is retained and that the quotient is not rounded off.**

* Use your notebook to develop an implementation of this formula.
* Then create a Python script called ``JD.py`` that asks a user to enter day, month and year. For this date, the script calculates the Julian Date (JD) and prints it. The day can be entered as a floating point number to allow for fractional days. 
* The article provides a test case: the Julian Date for January 4, 1700 is 2341975.5.    
  Use your code to get the correct result, and attach your screen shot in this notebook.&nbsp;

Extend the program on the notebook, so that it can calculate the number of days between two dates in the Gregorian calendar. What is your age expressed in days?
      
**Answers**:

**First version:**

````python
# Input variables for the Gregorian date

M = int(input("Enter month:"))
D = float(input("Enter day:"))
Y = int(input("Enter year:"))

# Compute JD for the Gregorian date

JD = 367*Y -7*int((Y+int((M+9)/12))/4) - 3*int((int((Y+int((M-9)/7))/100) + 1)/4) + int((275*M)/9) + D + 1721029-0.5
print(f"JD = {JD}")
````

**Extended version:**

````python
# Input variables for the first Gregorian date

M1 = int(input("Enter month 1:"))
D1 = float(input("Enter day 1:"))
Y1 = int(input("Enter year 1:"))

# Input variables for the second Gregorian date

M2 = int(input("Enter month 2:"))
D2 = float(input("Enter day 2:"))
Y2 = int(input("Enter year 2:"))

# Compute JD for the first Gregorian date

JD1 = 367*Y1 -7*int((Y1+int((M1+9)/12))/4) - 3*int((int((Y1+int((M1-9)/7))/100) + 1)/4) + int((275*M1)/9) + D1 + 1721029-0.5

# Compute JD for the second Gregorian date

JD2 = 367*Y2 -7*int((Y2+int((M2+9)/12))/4) - 3*int((int((Y2+int((M2-9)/7))/100) + 1)/4) + int((275*M2)/9) + D2 + 1721029-0.5

# Compute absolute difference between the two JD's

difference = abs(JD1 - JD2)
print(f"Number of days between the two dates = {difference}")
````

###  Calculating the roots of a second degree polynomial (3 pt) <font color='red'><b>COURSEWORK</b></font>

Use your notebook to develop the necessary code and then which solves for the roots
of $f(x)=ax^2+bx+c$ for values of ``a``, ``b`` and ``c`` entered by a user of your program.
All the required information related to this question is included in the notebook about conditions.

You have to remember that 

$$f(x)=ax^2+bx+c$$

has two solutions:

$$x_1=\frac{-b + \sqrt{b^2-4ac}}{2a} \quad\text{and}\quad x_2=\frac{-b - \sqrt{b^2-4ac}}{2a}$$

only if:

$$\begin{split}D= b^2-4ac>0\end{split}$$

It has one solution if $D == 0$ and no solutions (at least no real solutions) if $D < 0$.

* Use the ``if,elif,else`` construct to print either two solutions, one solution or an message that no solutions
 could be found. 
* Use f-strings for your output if there are solutions.   
* Include your code with an input example and its output in your notebook.
* Use your notebook code to create a script called ``abc.py``.
* On Terminal, run your python script, screenshot the result, and attach the it here. 

**Answer**:

````python
from math import sqrt

# Coefficients of the quadratic equation

a = float(input("Enter a:"))
b = float(input("Enter b:"))
c = float(input("Enter c:"))

# Compute discriminant

D = b**2-4*a*c

if D > 0:
    # Two real solutions
    x1 = (-b+sqrt(D))/(2*a)
    x2 = (-b-sqrt(D))/(2*a)
    print(f"The solutions are x1 = {x1} and x2 = {x2}")
elif D == 0:
    # One real solution
    x1 = (-b+sqrt(D))/(2*a)
    print(f"The solution is x = {x1}")
else:
    # No real solution if D < 0
    print("No solutions could be found")
````

### Uniqueness in lists using the `set` object (2 pt) <font color='red'><b>COURSEWORK</b></font>

Given two lists with a number of gamma ray bursts per night for two different periods

```python
    night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
    night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]
```

Use Python sequence type `set` to create a **list**:

1. with values that are both in `night1` and `night2`.
1. with values that are the difference between `night1` and `night2`.
1. with values that contains all items from both sets, duplicates are excluded.
1. after an update of `night1` by removing all its elements that are not available in `night2`.

**Answers**:

````python
# Lists for night1 and night2

night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]

# Convert lists to sets

n1_set = set(night1)
n2_set = set(night2)

# 1. Intersection of night1 and night2

inters = set.intersection(n1_set, n2_set)

# Convert intersection set back to list

list_inters = list(inters)
print(f"1. {list_inters}")

# 2. Difference between night1 and night2

diff = set.difference(n1_set, n2_set)

# Convert difference set back to list

list_diff = list(diff)
print(f"2. {list_diff}")

# 3. Union of night1 and night2

sym_diff = n1_set.symmetric_difference(n2_set)

# Convert union set back to list

list_sym_diff = list(sym_diff)
print(f"3. {list_sym_diff}")

# 4. Update night1 by removing elements not in night2

n1_set.intersection_update(n2_set)
night1 = list(n1_set)
print(f"4. {night1}")
````

### Apply NumPy functions to data (2pt) <font color='red'><b>COURSEWORK</b></font>

* Create a NumPy array `x` with 100 angles in degrees between 0 and 360 degrees using ``np.linspace()``
* Calculate the ``np.sinc(x)`` values of ``x``, with ``x`` in radians. The definition of the sinc function is:

  $$\mathrm{sinc}(x) = \frac{\sin(\pi x)}{\pi x}$$

* Print the first 10 values of the result. 
  
**Answer**

````python
import numpy as np

# Create 100 angles in degrees between 0 and 360

x_deg = np.linspace(0,360,100)

# Convert angles from degrees to radians

x_rad = np.radians(x_deg)

# Compute sinc(x) values

f_sinc = np.sinc(x_rad) 
print(f"First 10 values of np.sinc(x_rad) = {f_sinc[:10]}")
````
