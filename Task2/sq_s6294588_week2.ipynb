**Question 2.12 : Quadrants:**

```python
from math import atan2, degrees

# input variables for the x and y coordinate

x = float(input("Enter x"))
y = float(input("Enter y"))

# Compute angle between the x axis and the line between origin and (x,y)

angle_rad = atan2(y, x) 
angle_deg = degrees(angle_rad)
print(f"Angle = {angle_deg}")

# Determine the quadrant based on the angle

if 0 < angle_deg < 90:
    quadrant = "I"
elif 90 < angle_deg < 180:
    quadrant = "II"
elif -90 < angle_deg < 0:
    quadrant = "IV"
elif -180 < angle_deg < -90:
    quadrant = "III"
else:
    quadrant = "on an axis"

print(f"It lies in quadrant {quadrant}")
```

The difference between atan and atan2 is that atan computes arctan(y/x) and returns values in (−pi/2, pi/2), so it cannot determine the correct quadrant. atan2 uses both x and y to determine the correct quadrant and returns values in (−pi, pi).

**Question 2.14 : Exercise with slice():**

```python
# Original list of values

X = [1.36, 1.46, 1.41, 2.12, 0.42, 0.98, 0.45, 1.97, -0.24, 2.18]

# Modify the list

X.append(0.51)
X.remove(2.12)

print(f"list = {X}")

# Create slice object: start at index 2, stop at -2, step size 2

indices = slice(2,-2, 2)

# Apply slicing to list

X_sliced = X[indices]

# Compute sum of sliced values

sum_X_sliced = sum(X_sliced)

print(f"sliced list = {X_sliced}")
print(f"sum sliced list = {sum_X_sliced}")
```

**Question 2.19 : First steps using NumPy arrays:**

```python
import numpy as np
mu, sigma = 0, 1.4 # mean and standard deviation

# Generate 100000 random values from a normal distribution

s = np.random.normal(mu, sigma, 100000)
# Compute standard deviation

s_d = abs(sigma - np.std(s, ddof=1))
print(f"standard deviation = {s_d}")
```

**Question 2.21 : Fibonacci sequence:**

```python
# Starting values of Fibonacci sequence

a = 1
b = 2

# Compute the 100th Fibonacci term

for i in range(99):
    c = a + b
    a = b
    b = c
print(f"100-th term = {a}")
```
