**Question 3.1 : Average calculator:**

```python
# List of masses

masses = [1.9891e+30, 1.8986e+27, 
          5.6846e+26, 1.0243e+26, 8.6810e+25,
          5.9736e+24, 4.8685e+24, 6.4185e+23, 
          3.3022e+23, 7.349e+22, 1.25e22]

# Only print the masses smaller or equal of the mass of the moon

new_masses = []
for i in masses:
    if i > masses[-2]:
        new_masses.append(i)
print(f"New masses = {new_masses}")

# Only print the last 5 masses

indices = slice(-5,None,1)
sliced_masses = masses[indices]
print(f"Last 5 masses = {sliced_masses}")

# Avererge of last 5 masses

sum_sliced_masses = sum(sliced_masses)
len_sliced_masses = len(sliced_masses)
print(f"Average of last 5 masses = {sum_sliced_masses / len_sliced_masses}")
```

**Question 3.2 : Calculate escape velocities:**

```python
from scipy.constants import G
from math import sqrt
masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

v_esc_list = []

# Compute escape velocity and print masses with 2 decimals

for i in range(len(names)): 
    m = masses[i]
    r = radii[i]
    v_esc = sqrt((2*G*m)/r)
    v_esc_list.append(v_esc)
    print(f"Escape velocity {names[i]} = {v_esc}")
    print(f"m = {m:6.2f}")

# Make the escape velocities a dictionary

dic = dict(zip(names, v_esc_list))

print(f"Escape velocity = {dic}")
```

**Question 3.5 : Convergence of a sequence:**

```python
N = 200
total = 0

for x in range(1, N + 1):
    sum = 1 / (x**2)
    total = total + sum

print(f"Sum from 1 to {N} = {total}")
```

**Question 3.7 : Angular size:**

```python
def angular_size(d, D):
    from math import atan, radians
    
    delta = 2*atan(d/(2*D))
    delta_degrees = radians(delta)
    return delta_degrees

print(f"Angular size = {angular_size(1,1)}")
```

**Question 3.10 : Area and volume of a cylinder:**

```python
def cylinder(r, h):
    from math import pi
    
    A = 2*pi*r**2+2*pi*r*h
    V = pi*r**2*h
    return A, V

area, volume = cylinder(1, 1)
ratio = area / volume
print(f"Area = {area}")
print(f"Volume = {volume}")
print(f"Area/volume = {ratio}")
```

**Question 3.17 : Standard deviation using ndarray and in a one-liner:**

```python
import numpy as np

# Define standard deviation function

def stdev(y):
    N = len(y)
    mu = np.sum(y) / N
    sigma = np.sqrt((np.sum((y-mu)**2)) / N)
    print(sigma)

# Test function with test array

y = np.array([1, 2, 3, 4, 5])
stdev(y)
print(np.std(y))
```

**Question 3.18 : Least Squares fitting:**

```python
import numpy as np

# Define data and compute averages

data = np.array([
    [18.49, 2094.75],
    [35.48, 2320.87],
    [35.48, 1224.9],
    [11.34, 872.13],
    [7.12,  432.2],
    [9.82,  751],
    [23.66, 1410],
    [44,   2700],
    [66.7,  5292],
    [52.4,  2550],
    [55,    5253],
    [25.6,  2419],
    [26.19, 2158]
])
x = data[:, 0]
y = data[:, 1]
x_average = np.average(x)
y_average = np.average(y)

# Compute least square fitting

b = (sum((x - x_average) * (y - y_average)))/(sum((x - x_average)**2))
a = y_average - b * x_average
print(b)
print(a)
```

**Question 3.22 : Plotting a catenary:**

```python
from math import cosh
import numpy as np
from matplotlib import pyplot as plt

# Compute x and y values with range

x_values = list(range(-5, 6))
y_values = []
for x in x_values:
    y = cosh(x)
    y_values.append(y)

# Compute x and y values with numpy

x_values_np = np.arange(-5, 6, 1)
y_values_np = np.cosh(x_values_np)

# Plot x and y values

plt.plot(x_values, y_values, label="cosh(x)")
plt.plot(x_values_np, y_values_np, label="cosh(x) np")
plt.grid()
plt.title("Catenary", fontsize=15)
plt.xlabel("x", fontsize=15)
plt.ylabel("cosh(x)", fontsize=15)
plt.xticks(range(-5, 6, 1))
plt.yticks(range(0, 81, 10))
plt.legend()
plt.show()
```
