**Question 6.1 : FITS file analysis**

```python
from astropy.io import fits
hdulist = fits.open("m101.fits")
hdulist.info()
```

It has 2 HDU's. The first one: index 0. Number type = int16

```python
hdr = hdulist[0].header
for key in hdr[:5]:
    print(f"{key:18s}= {hdr[key]}")
```

Yes it does, 16 = int16

```python
dat = hdulist[0].data
shape = dat.shape
dtype = dat.dtype
print(f"The shape is {shape}")
print(f"The type {dtype}")
```

It does correspond to info(). i stands for integer and it is 2 bytes so 16 bits so int16

```python
from scipy import stats
import numpy as np
nobs, (minval, maxval), mean, var, skew, kurt = stats.describe(dat.flat)
print(f"N={nobs} min={minval}, max={maxval}, mean={mean}, std={var**0.5}")

mode, counts = stats.mode(dat.flat)
print(mode, counts)
print(f"Mode of the data: {mode} with {counts} counts")

median = np.median(dat.flat)
print(f"Median of the data: {median}")
```

```python
from matplotlib.pyplot import figure, show
fig = figure()
frame = fig.add_subplot(1,1,1)
frame.hist(dat.flatten(), bins=50, color="orange")
frame.grid()
frame.set_xlabel("Pixel value")
frame.set_ylabel("N")
frame.axvline(mean, color='r', label = "Mean" )
frame.axvline(median, color='b', label = "Median")
frame.axvline(mode, color='g', label = "Mode")
frame.legend()
show()
```

```python
import warnings
from astropy.io.fits.verify import VerifyWarning
warnings.simplefilter('ignore', VerifyWarning)

data = np.asarray(dat, dtype=float)
median = np.median(data)
fig = figure()
frame = fig.add_subplot()
frame.set_xlabel(hdr['CTYPE1'])
frame.set_ylabel(hdr['CTYPE2'])
mappable = frame.imshow(data, vmin = median, origin = "lower")
cbar = fig.colorbar(mappable)
cbar.set_label("Counts")
show()

hdr["BUNIT"] = "Counts"
hdulistout = fits.PrimaryHDU(data)
hdulistout.header = hdr
hdulistout.writeto('m101_Counts.fits', overwrite=True, output_verify='ignore')
```

```python
import matplotlib.pyplot as plt
slice_image = dat[0:20, 0:20]
RMS = np.sqrt(np.mean(slice_image**2))
mean = np.mean(slice_image)
std = np.std(slice_image)
print(f"RMS = {RMS}")
print(f"Mean = {mean}")
print(f"Std = {std}")
```

The RMS spreads around 0 and the std spreads around the mean. The std quantifies noise the best, because it is the average fluctuations of pixels from the mean and RMS is the total noise

An image read with astropy.io.fits is stored as a NumPy array. A table is stored as a structured array, which contains rows and named columns similar to a spreadsheet. The image is accessed using array indices, while a table is accessed by row index or by column name.

```python
from astropy.io import fits
from astropy.table import Table

# Open the FITS file
hdul = fits.open('APOGEE_stars.fits')

# Access the primary data (in this case, it is a Table)
data = hdul[1].data  # Often, the first extension contains the table
colnames = hdul[1].columns.names # List of the column names

table = Table(data)

mask = (
    (data['STARFLAG'] == 0) &
    (data['ASPCAPFLAG'] == 0) &
    (data['SNR'] > 10) &
    (data['GAIAEDR3_PARALLAX'] / data['GAIAEDR3_PARALLAX_ERROR'] > 5) &
    (~np.isnan(data['GAIAEDR3_PHOT_G_MEAN_MAG'])) &
    (~np.isnan(data['GAIAEDR3_PHOT_BP_MEAN_MAG'])) &
    (~np.isnan(data['GAIAEDR3_PHOT_RP_MEAN_MAG']))
)

table['DATAFLAG'] = mask.astype(int)
print(table['DATAFLAG'])
```
